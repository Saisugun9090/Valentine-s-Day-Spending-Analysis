```python
decade_spending['PerPersonSpendingChange'] = decade_spending['PerPersonSpending'].pct_change() * 100
decade_spending['CandyChange'] = decade_spending['Candy'].pct_change() * 100
decade_spending['FlowersChange'] = decade_spending['Flowers'].pct_change() * 100
decade_spending['JewelryChange'] = decade_spending['Jewelry'].pct_change() * 100
decade_spending['GreetingCardsChange'] = decade_spending['GreetingCards'].pct_change() * 100
decade_spending['EveningOutChange'] = decade_spending['EveningOut'].pct_change() * 100
decade_spending['ClothingChange'] = decade_spending['Clothing'].pct_change() * 100
decade_spending['GiftCardsChange'] = decade_spending['GiftCards'].pct_change() * 100
```

```python
changes = decade_spending[['Year', 'PerPersonSpendingChange', 'CandyChange', 'FlowersChange', 'JewelryChange',
                                'GreetingCardsChange', 'EveningOutChange', 'ClothingChange', 'GiftCardsChange']]
print("\nYear-to-Year Percentage Changes:")
print(changes)
```

```python
plt.figure(figsize=(12, 8))
sb.lineplot(data=changes, x='Year', y='PerPersonSpendingChange', marker='o', label='PerPersonSpending')
plt.title('Year-to-Year Percentage Change in Per Person Spending')
plt.xlabel('Year')
plt.ylabel('Percentage Change')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sb.lineplot(data=changes, x='Year', y='CandyChange', marker='o', label='Candy')
plt.title('Year-to-Year Percentage Change in Candy Spending')
plt.xlabel('Year')
plt.ylabel('CandyChange')
plt.legend()
plt.show()
```

```python

average_changes = decade_spending[['CandyChange', 'FlowersChange', 'JewelryChange', 'GreetingCardsChange',
                                      'EveningOutChange', 'ClothingChange', 'GiftCardsChange']].mean()

print("\nAverage Year-to-Year Percentage Changes for Each Item:")
print(average_changes)

plt.figure(figsize=(12, 8))
average_changes.plot(kind='bar', color='skyblue')
plt.title('Average Year-to-Year Percentage Changes for Each Item')
plt.xlabel('Item')
plt.ylabel('Average Percentage Change')
plt.show()
```
